```python
import string
import random
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
```

# Generate samples 

```python
ordered = True

big_set_cars = [a for a in string.ascii_uppercase]
for a in string.ascii_uppercase:
    for b in string.ascii_uppercase:
        for c in string.ascii_uppercase:
            big_set_cars.append(a+b+c)

ranges = [200,400,600,800,1000]
for range in ranges: 
    with open('../examples/document/to_send/test4_'+str(range)+'.txt','w') as f:
        cars = big_set_cars[:range]
        if not ordered:
            random.shuffle(cars)
        f.write('NumLines=2;' + '\nMaxColumns=' + str(len(cars)+1) + ';')
        all_cars = '.'.join(a for a in cars)
        f.write('\nBlocks='+all_cars + ';')
        initialState = ''
        initialState += 'LastDock('+cars[0]+')'
        for i, car in enumerate(cars): 
            if i < len(cars)-1:
                initialState += '.' + 'NextToDock(' + car + ',' + cars[i+1] + ')'
        initialState += '.FirstDock('+cars[-1]+')'
        f.write('\nInitialState=' + initialState+ ';')
        if not ordered:
            random.shuffle(cars)
        goalState = ''
        goalState += 'FirstFerry('+cars[0]+')'
        for i, car in enumerate(cars): 
            if i < len(cars)-1:
                goalState += '.' + 'NextToFerry(' + cars[i+1] + ',' + car + ')'
        goalState += '.LastFerry('+cars[-1]+')'
        f.write('\nGoalState=' + goalState + ';')
```
