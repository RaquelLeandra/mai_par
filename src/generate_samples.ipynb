```python
import string
cars = [a for a in string.ascii_uppercase]
for a in string.ascii_uppercase:
    for b in string.ascii_uppercase:
        cars.append(a+b)
print('NumLines=2;','\nMaxColumns=' + str(len(cars)+1) + ';')
all_cars = '.'.join(a for a in cars)
print('\nBlocks='+all_cars + ';')
initialState = ''
initialState += 'LastDock('+cars[0]+')'
for i, car in enumerate(cars): 
    if i < len(cars)-1:
        initialState += '.' + 'NextToDock(' + car + ',' + cars[i+1] + ')'
initialState += '.FirstDock('+cars[-1]+')'
print('\nInitialState=' + initialState+ ';')

goalState = ''
goalState += 'FirstFerry('+cars[0]+')'
for i, car in enumerate(cars): 
    if i < len(cars)-1:
        goalState += '.' + 'NextToFerry(' + cars[i+1] + ',' + car + ')'
goalState += '.LastFerry('+cars[-1]+')'
print('\nGoalState=' + goalState + ';')
```
